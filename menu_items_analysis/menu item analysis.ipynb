```python
import pandas as pd
```

```python
start = pd.Timestamp('2025-03-01').date()
end = pd.Timestamp('2025-04-01').date()

tmis = (pd
        .read_csv('assets/lido_transaction_menu_items.csv',
                  parse_dates=['CreateDate'],
                  date_format='%Y-%m-%d %H:%M:%S',
                  usecols=['Store', 'Price', 'MenuItemId', 'CreateDate'],
                  dtype={'Store': 'category', 'Price': 'float64', 'MenuItemId': 'int64'})
        .dropna(how='all'))
tmis = tmis.loc[tmis['CreateDate'].dt.date.between(start, end)]
tmis = tmis.reset_index(drop=True)
tmis['CreateDate'] = tmis['CreateDate'].dt.date
tmis
```

```python
tmis.info()
```

```python
 fives = tmis['MenuItemId'] == 5
 erods = tmis['Store'] == 'Erod'
 march1 = tmis['CreateDate'] == pd.Timestamp('2025-03-01').date()
 tmis[fives & erods & march1]
```

```python
grouped = tmis.groupby([tmis['CreateDate'], 'Store', 'MenuItemId'], observed=False)
grouped = grouped.agg({
    'MenuItemId': 'count'
})
grouped.rename(columns={'MenuItemId': 'Count'}, inplace=True)
grouped
```
